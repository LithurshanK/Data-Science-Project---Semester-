# Filling 'Year_of_Birth' missing values

```python
# visualize 'Year_of_Birth' with its count for 'Single' and "married" with different plots
# use countplot
sns.countplot(x='Year_of_Birth', data=chatterbox[chatterbox['Marital_Status'] == 'Single'])
plt.xlabel('Year of Birth')
plt.ylabel('Count')
plt.title('Count Plot: Year of Birth for Single')
plt.show()
```

```python
# visualize 'Year_of_Birth' with its count
sns.countplot(x='Year_of_Birth', data=chatterbox,hue='Marital_Status')
plt.xticks(rotation=90)
plt.show()
```

if 'Year_of_Birth' is a discrete or categorical variable, or if it has a skewed distribution, using the mode could be a better option. Ultimately, the decision depends on your specific data and the characteristics of the 'Year_of_Birth' variable.

```python
# Calculate the mode of 'Year_of_Birth' for 'Marital_Status' being single
single_mode = chatterbox[chatterbox['Marital_Status'] == 'Single']['Year_of_Birth'].mode().values[0]

# Calculate the mode of 'Year_of_Birth' for 'Marital_Status' being married
married_mode = chatterbox[chatterbox['Marital_Status'] == 'Married']['Year_of_Birth'].mode().values[0]

# Fill missing values of 'Year_of_Birth' based on 'Marital_Status'
chatterbox.loc[(chatterbox['Marital_Status'] == 'Single') & (chatterbox['Year_of_Birth'].isnull()), 'Year_of_Birth'] = single_mode
chatterbox.loc[(chatterbox['Marital_Status'] == 'Married') & (chatterbox['Year_of_Birth'].isnull()), 'Year_of_Birth'] = married_mode
```

# Filling 'Married Status' missing values

```python
# Group by 'Year_of_Birth' and 'Marital_Status' and select the mode
marital_status_mode = chatterbox.groupby(['Year_of_Birth'])['Marital_Status'].apply(lambda x: x.mode().iloc[0])

# Fill missing values in 'Marital_Status' with the mode based on 'Year_of_Birth'
chatterbox['Marital_Status'] = chatterbox['Marital_Status'].fillna(chatterbox['Year_of_Birth'].map(marital_status_mode))
```

```python
chatterbox.Year_of_Birth.isnull().sum()
```

```python
sns.violinplot(x='Marital_Status', y='Year_of_Birth', data=chatterbox)
plt.xlabel('Marital Status')
plt.ylabel('Year of Birth')
plt.title('Violin Plot: Year of Birth by Marital Status')
plt.show()
```

```python
# fill missing values in 'Marital_Status' with mode
chatterbox['Marital_Status'] = chatterbox['Marital_Status'].fillna(chatterbox['Marital_Status'].mode().values[0])

# fill missing values in 'Year_Of_Birth' with integer value median value
chatterbox['Year_of_Birth'] = chatterbox['Year_of_Birth'].fillna(int(chatterbox['Year_of_Birth'].median()))
```

```python
sns.violinplot(x='Marital_Status', y='Year_of_Birth', data=chatterbox)
plt.xlabel('Marital Status')
plt.ylabel('Year of Birth')
plt.title('Violin Plot: Year of Birth by Marital Status')
plt.show()
```

```python
# if Inactive_Date is 0000-00-00 and Status is Active change Inactive_Date to '\N'
chatterbox.loc[(chatterbox['Inactive_Date'] == '0000-00-00') & (chatterbox['Status'] == 'Active'), 'Inactive_Date'] = '\\N'
```

```python
# if Date_Resigned is 0000-00-00 and Status is Active change Date_Resigned to '\N'
chatterbox.loc[(chatterbox['Date_Resigned'] == '0000-00-00') & (chatterbox['Status'] == 'Active'), 'Date_Resigned'] = '\\N'
```

```python
# if Inactive_Date is not 0000-00-00 and Date_Resigned is 0000-00-00 fill with Inactive_Date
chatterbox.loc[(chatterbox['Inactive_Date'] != '0000-00-00') & (chatterbox['Date_Resigned'] == '0000-00-00'), 'Date_Resigned'] = chatterbox['Inactive_Date']
```

```python
# Change title to 'Mr' for male employees with titles 'Ms' or 'Miss'
chatterbox.loc[(chatterbox['Gender'] == 'Male') & (chatterbox['Title'].isin(['Ms', 'Miss'])), 'Title'] = 'Mr'
```

```python
import random

# Define a function to randomly assign title
def assign_random_title():
    return random.choice(['Ms', 'Miss'])

# Change title to 'Ms' or 'Miss' randomly for female employees with title 'Mr'
female_mr_indices = (chatterbox['Gender'] == 'Female') & (chatterbox['Title'] == 'Mr')
chatterbox.loc[female_mr_indices, 'Title'] = chatterbox.loc[female_mr_indices].apply(lambda row: assign_random_title(), axis=1)
```

```python
# if Reporting_emp_1 is same as Employee_No change Reporting_emp_1 to '\N' where Reporting_emp_1 is object and Employee_No is int64

chatterbox.loc[chatterbox['Reporting_emp_1'] == chatterbox['Employee_No'], 'Reporting_emp_1'] = '\\N'
```
